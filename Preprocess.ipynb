```python
import numpy as np
import pandas as pd
import string
```

```python
data = pd.read_csv('./dataset/lyrics.csv')
```

```python
filtered = data[data['lyrics'].notnull() & data['artist'].notnull()]
filtered
```

```python
cleaned = filtered.copy()
cleaned['lyrics'] = cleaned['lyrics'].str.replace("[-\?.,\/#!$%\^&\*;:{}=\_~()]", ' ')
cleaned['lyrics'] = cleaned['lyrics'].str.replace("\[(.*?)\]", ' ') # remove things like [Chorus] or [Verse]
cleaned['lyrics'] = cleaned['lyrics'].str.replace("' | '", ' ')
cleaned['lyrics'] = cleaned['lyrics'].str.replace('x[0-9]+', ' ')
cleaned['lyrics'] = cleaned['lyrics'].str.replace('\s+', ' ')
# Could remove chorus/verse/etc even if they don't appear in lyrics
cleaned
```

```python
cleaned = cleaned[cleaned['lyrics'].str.strip().str.lower() != 'instrumental']
print(cleaned.shape)
cleaned = cleaned[~cleaned['lyrics'].str.contains(r'[^\x00-\x7F]+')]
```

```python
cleaned.to_csv('./dataset/cleaned_dataset.csv')
```
